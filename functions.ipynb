```python
def eligible(gender,age):
    if(gender=="male"):
        if(age>=21):
            return "elgible"
        else:
            return "not eligible"
    elif(gender=="female"):
        if(age>=18):
            return "eligible"
        else:
            return "not eligible"
    else:
        return "invalid gender"
gender=input("enter ur gender")
age=int(input("enter ur age"))
result=eligible(gender,age)
print(result)
```

```python
def areaoftriangle(height,breadth):
    area=height*breadth/2
    return area
def perimeteroftriangle(height1,height2,breadth1):
    perimeter=height1+height2+breadth1
    return perimeter
   
height=int(input("enter height:"))
breadth=int(input("enter breadth:"))
height1=int(input("enter height1:"))
height2=int(input("enter height2:"))
breadth1=int(input("enter breadth:"))
area=areaoftriangle(height,breadth)
perimeter=perimeteroftriangle(height1,height2,breadth1)
print("area of triangle:",area)
print("perimeter of triangle:",perimeter)

```
